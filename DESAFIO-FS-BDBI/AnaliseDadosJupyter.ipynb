```python

import pandas as pd
```

```python
import matplotlib.pyplot as plt
from tabulate import tabulate
```

```python
df = pd.read_csv("//content/dados_rh.csv")
```

```python
df
```

```python
num_linhas, num_colunas = df.shape

print("Número de linhas:", num_linhas)
print("Número de colunas:", num_colunas)
```

```python
dados_nulos = df.isnull().sum()
print(dados_nulos)
print(dados_nulos[dados_nulos > 0])
```

```python

apuracao = df.groupby('Department')['EmployeeNumber'].count()
apuracao.plot(kind='bar')
plt.title('Quantidade de Funcionários por Setor')
plt.xlabel('Setor')
plt.ylabel('Número de Funcionários')
plt.xticks(rotation=60)
```

```python
apuracao = df.groupby('Education')['EmployeeNumber'].count()
apuracao.plot(kind='line')
plt.title('Quantidade de Funcionários por Nível de Escolaridade')
plt.xlabel('Nível de Escolaridade')
plt.ylabel('Número de Funcionários')
plt.show()
```

```python
apuracao = df.groupby('Attrition')['EmployeeNumber'].count()
apuracao.plot(kind='pie', autopct='%1.1f%%')
plt.title('Percentual de Funcionários com e sem Atrito')
plt.ylabel('')
plt.show()
```
