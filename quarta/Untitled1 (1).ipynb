```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
from tabulate import tabulate
```

```python
df = pd.read_csv("/content/dados (1).csv")
```

```python
df
```

```python
df2 = df.groupby('test preparation course')['math score'].agg(['min', 'max', 'mean'])
df2
print(tabulate(df2, headers='keys', tablefmt='fancy_grid'))
```

```python
df3 = df.groupby('test preparation course')[['reading score']].agg(['min', 'max', 'mean'])
df3
print(tabulate(df3, headers='keys', tablefmt='fancy_grid'))
```

```python
df4 = df.groupby('test preparation course')[['writing score']].agg(['min', 'max', 'mean'])
df4

print(tabulate(df4, headers='keys', tablefmt='fancy_grid'))
```

```python
df_grouped = df.groupby('race/ethnicity')['math score'].mean()

plt.bar(df_grouped.index, df_grouped.values)
plt.xticks(rotation=80)
plt.xlabel('Grupos de raça/etnia')
plt.ylabel('Média das notas em matemática')
plt.title('Média das notas em matemática por raça/etnia')
plt.show()
```

```python
df_grouped = df.groupby('parental level of education')['writing score'].mean()
plt.bar(df_grouped.index, df_grouped.values)
plt.xticks(rotation=80)
plt.xlabel('Nível de escolaridade dos pais')
plt.ylabel('Média das notas em escrita')
plt.title('Média das notas em escrita por nível de escolaridade dos pais')
plt.show()
```

```python
df_grouped = df.groupby('test preparation course')['writing score'].mean()
plt.bar(df_grouped.index, df_grouped.values)

plt.ylabel('Média das notas de escrita')
plt.title('Média das notas em escrita de acordo com o curso de preparaçao')
plt.xticks(rotation=45)
plt.show()



```
